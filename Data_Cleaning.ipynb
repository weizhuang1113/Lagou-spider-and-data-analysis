**爬虫内容包含许多无意义字段，后续进行删除处理**

**去除无用字段**

```python
to_drop = ['pcShow','appShow','deliver','score','approve','companyLogo','publisherId','hitags','formatCreateTime','adWord','resumeProcessRate','resumeProcessDay','imState','lastLogin','explain','plus','pcShow','appShow','deliver','gradeDescription','promotionScoreExplain']
df_noNeed = df.drop(to_drop,axis=1)
df_noNeed
```

**去除重复值**

```python
#positionId 为唯一值
len(df_noNeed['positionId'].unique())
```

```python
df_noDup = df_noNeed.drop_duplicates(subset = 'positionId')
```

**salary字段处理**

```python
df_noDup['salary'].unique()
```

```python
def cut_salary(salary,method):
    try:
        salary_list = salary.split('-')
        bottomSalary = int(salary_list[0][:-1])
        topSalary = int(salary_list[1][:-1])
    except:
        bottomSalary  = int(salary.upper().split('K')[0])
        topSalary = bottomSalary 
    if method == 'bottom':
        return bottomSalary
    else:
        return topSalary
```

```python
df_noDup['topSalary'] = df_noDup['salary'].apply(cut_salary,method = 'top')
```

```python
df_noDup['bottomSalary'] = df_noDup['salary'].apply(cut_salary,method = 'bottom')
```

```python
df_noDup['avgSalary'] = df_noDup.apply(lambda x: (x.topSalary + x.bottomSalary)/2,axis=1)
```

```python
df_noDup['crateTime'] = pd.to_datetime(df_noDup['createTime'])
print(len(df_noDup))
print(df_noDup['crateTime'].max())
print(df_noDup['crateTime'].min())
```
